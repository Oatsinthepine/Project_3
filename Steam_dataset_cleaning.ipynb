```python
#import dependencies
import os
import pandas as pd
import numpy as np
```

```python
# reading both csv files and convert to data frame
file_path_1 = os.path.join('Resource', 'games.csv')
file_path_2 = os.path.join('Resource', 'steam.csv')
df = pd.read_csv(file_path_1, encoding='utf-8')
df.head(5)
```

```python
supplement_df = pd.read_csv(file_path_2, encoding='utf-8')
supplement_df.head()
```

```python
#check column names
df.columns
```

```python
# select only required columns for analysis:
new_df = df[['AppID','Name','Release date','Peak CCU','Required age','Price','Windows',
       'Mac', 'Linux','Metacritic score','Positive','Negative','Achievements','Average playtime forever','Average playtime two weeks',
       'Median playtime forever', 'Median playtime two weeks', 'Developers',
       'Publishers', 'Categories', 'Genres']]

new_df.head()
```

```python
# convert the release date into datetime
new_df['Release_date'] = pd.to_datetime(new_df['Release date'], errors='coerce')
new_df.info()
```

```python
new_df['Release_date'].head()
```

```python
# sort the dataframe based on release date from earilest to most recent.
new_df_sorted = new_df.sort_values(by='Release_date', ascending=True)
new_df_sorted.head()

# so we can see it contains many really old games data, let just take the data for 10 years time intervals from 2014 to 2024.
```

```python
# drop the original release date column as no-longer needed
new_df_sorted = new_df_sorted.drop('Release date', axis=1)
```

```python
# removeing very old games
new_df_selected = new_df_sorted[(new_df_sorted['Release_date'] >= '2014-01-01') & (new_df_sorted['Release_date'] <= '2024-12-31')]
new_df_selected.tail()
```

```python
#check both df's columns
print(supplement_df.columns)
print(new_df_selected.columns)
```

```python
#selecting the required columns from supplement df and rename the columns for later merging.
new_df_selected = new_df_selected[['AppID', 'Name', 'Release_date', 'Peak CCU', 'Developers', 'Publishers','Windows', 'Mac',\
       'Linux', 'Required age','Categories', 'Genres','Achievements','Metacritic score', 'Positive', 'Negative',\
        'Average playtime forever', 'Average playtime two weeks','Median playtime forever', 'Median playtime two weeks',\
        'Price']]

new_df_selected = new_df_selected.rename(columns={
    'Positive': 'Positive_ratings',
    'Negative': 'Negative_ratings'
})
```

```python
# here just checking how  many missing values in the main (games.csv) playtime columns
print(len(new_df_selected[new_df_selected['Average playtime forever']==0]))
print(len(new_df_selected[new_df_selected['Average playtime two weeks'] == 0]))
print(len(new_df_selected[new_df_selected['Median playtime forever']==0]))
print(len(new_df_selected[new_df_selected['Median playtime two weeks']==0]))

# it seems when gathering raw data that the playtime columns are missing many values. Thus we need to merge it with the supplement df.
```

```python
check_df_1 = new_df_selected[((new_df_selected['Average playtime forever']==0) & (new_df_selected['Median playtime forever'] == 0))]
check_df_2 = new_df_selected[(new_df_selected['Average playtime two weeks']==0) & (new_df_selected['Median playtime two weeks']==0)]   
#print(len(check_df_1))
print(len(check_df_2))

# in the games.csv in this raw data there are a lot missing data for two weeks, I decide to drop them and only keep the forever time for average and median,
# after this I will merge this two column with the playtime from steam.csv to supplement some missing data.
```

```python
new_df_selected = new_df_selected.drop(['Average playtime two weeks', 'Median playtime two weeks'], axis = 1)
new_df_selected
```

```python
new_df_selected = new_df_selected.rename(columns={
    'Average playtime forever': 'Average playtime',
    'Median playtime forever': 'Median playtime'
})

print(new_df_selected.columns)
```

```python
supplement_df_subset = supplement_df[['appid', 'name', 'release_date','developer', 'publisher','required_age', 'categories', 'genres',\
                                      'achievements', 'positive_ratings', 'negative_ratings','average_playtime', 'median_playtime',\
                                        'price']]

# Renaming the suplment df and prepare for merging
supplement_df_subset = supplement_df_subset.rename(columns={
    'appid':'AppID',
    'name':'Name',
    'release_date':'Release_date',
    'developer':'Developers',
    'publisher': 'Publishers',
    'required_age':'Required age',
    'categories':'Categories',
    'genres':'Genres',
    'achievements':'Achievements',
    'positive_ratings':'Positive_ratings',
    'negative_ratings':'Negative_ratings',
    'average_playtime':'Average playtime',
    'median_playtime': 'Median playtime',
    'price':'Price'
})

print(supplement_df_subset.columns)
```

```python
#sorting supplement_df before merge
supplement_df_subset = supplement_df_subset.sort_values(by='Release_date', ascending=True)
```

```python
# also selecting the wanted time-frame game data before merge.
supplement_df_subset_selected = supplement_df_subset[(supplement_df_subset['Release_date'] >= '2014-01-01') & \
                                                    (supplement_df_subset['Release_date'] <= '2024-12-31')]
supplement_df_subset_selected
```

```python
# convert to datetime
supplement_df_subset_selected['Release_date'] = pd.to_datetime(supplement_df_subset_selected['Release_date'], errors='coerce')
```

```python
#merging two df together, I use Left join as the new_df_selected is the main df I want to retain, supplement_df will be joined to added any missing data.

'''Because the steam.csv and games.csv are actually contains many similar data, they both obtained via the original auther used steamspy,
so after thoroughly checking and comparing these two dataframes, as in both csv files, the appid, games name etc are the same, Therefore, two dfs can be merged, and in this way, some missing values 
can be filled for better analysis'''

merged_df = pd.merge(new_df_selected, supplement_df_subset_selected, how='left',\
                    on=['AppID', 'Name', 'Release_date', 'Developers', 'Publishers',
       'Required age', 'Categories', 'Genres', 'Achievements',
       'Positive_ratings', 'Negative_ratings', 'Average playtime',
       'Median playtime', 'Price'])
```

```python
#checking after merge.
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.isna().sum()
```

```python
# remove all NaN after mergering
merged_df = merged_df.dropna(how='any')
merged_df.head()
```

```python
# further clean-up, now we have the merged_df but as we inspect there are a lot of rows with multiple columns value equal to 0, I will select 
# subset that if both potive and negative rating that are not equal to 0.
merged_df_selected = merged_df.loc[~((merged_df['Positive_ratings']==0) & (merged_df['Negative_ratings']==0))]
merged_df_selected
```

```python
# further slecting from the subset that choose only rows and all the columns that if not all peak_ccu, metacritic score, average and median playtime are equal to 0
merged_df_selected_further = merged_df_selected.loc[~((merged_df_selected['Peak CCU']==0) & (merged_df_selected['Metacritic score']==0) &\
                                                      (merged_df_selected['Average playtime']==0) & \
                                                      (merged_df_selected['Median playtime']==0))]

# final cleaned and selected version of data frame that ready for further analysis.
merged_df_selected_further = merged_df_selected_further.rename(columns={
    'Peak CCU': 'Peak_CCU',
    'Required age':'Required_age',
    'Metacritic score': 'Metacritic_score',
    'Average playtime':'Average_playtime',
    'Median playtime':'Median_playtime'
})                                          
```

```python
merged_df_selected_further.tail()
```

```python
# export the final data frame as output file for later use.
merged_df_selected_further.to_csv('cleaned_steam_data.csv', index=False, header=True, encoding='utf-8')
```
